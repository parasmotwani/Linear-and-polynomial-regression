# Linear regression

```python
x = np.random.normal(3.0, 1.0, 1000)
y = 100 - (x + np.random.normal(0,0.1,1000))*3
plt.scatter(x, y)
```

```python
plt.scatter(x, y)
def predit(x):
    return -3*x + 100
fitline = predit(x)
plt.plot(x, fitline, c = 'b')
plt.show()
```

np.random.seed(4)

# Define a polynomial relationship with some noise
# y = ax^2 + bx + c + noise
a, b, c = 0.02, -1.5, 10  # Coefficients for the polynomial
noise = np.random.normal(0, 15.0, 10000)  # Add noise to y
y = a * x**2 + b * x + c + noise

```python
np.random.seed(4)
y = np.random.normal(50.0, 10.0, 1000)/x
x = np.random.normal(3.0, 1.0, 1000)
```

```python
plt.scatter(x, y)
```

```python
a = np.array(x)
b = np.array(y)

p2 = np.poly1d(np.polyfit(a,b,9))
xp = np.linspace(0,6,3000)
plt.scatter(x, y)
plt.plot(xp, p2(xp), c = 'b')
plt.show()
```
